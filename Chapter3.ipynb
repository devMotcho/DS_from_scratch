#### Data Visualization with matplotlib

#### Line Chart

```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# ciar um line chart, years no x-axis e gpd no y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# adicionar titulo
plt.title('Norminal GDP')

# adicionar etiqueta ao y-axis
plt.ylabel('Billions of $')
plt.show()
```

#### Bar Charts

```python
# Bom para mostrar a diferena entre quantidades entre um discreto conjunto de itens.

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

#bars sao por default width 0.8, adicionamos 0.1 para as coordenadas da esquerda
# para que cada barra esteja centrada.
xs = [i + 0.1 for i, _ in enumerate(movies)]

plt.bar(xs, num_oscars)

plt.ylabel('# of Academy Awards')
plt.title('My favorite Movies')

# labels x-axis com o nome dos filmes:
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
# um grafico de barras tambem  uma boa escolha para mostrar histogramas
# de buckets numericos m para visualmente explorar como os valores sao distribuidos
from collections import Counter
import matplotlib.pyplot as plt
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([(x - 4) for x in histogram.keys()], # movimenta cada barra para a esquerda em 4
        histogram.values(), #d a cada barra a sua altura correta
        8) # d a cada barra largura 8
plt.axis([-5, 105, 0, 5]) # x de -5 a 105 e y de 0 a 5
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of Students')
plt.title('Distribution of Exam 1 Grades')
plt.show()
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x + y for x, y in zip(variance,bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9) # top center
plt.xlabel('model complexity')
plt.title('The Bias-variance Tradeoff')
plt.show()
```

#### Scatterplots - Grafico de Disperso

```python
# Bom para visualizar relao entre 2 sets de dados
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)

# dar nome a cada ponto
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy = (friend_count, minute_count),
                 xytext = (5, -5),
                 textcoords ='offset points')
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```
